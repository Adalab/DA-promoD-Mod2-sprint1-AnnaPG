> Se quiere estudiar si existe una relaci칩n entre la cantidad de visitas que reciben los videos de m칰sica de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. 

Por lo tanto, querremos contestar a la siguiente pregunta: 

> 쯉on aquellos artistas que tienen m치s visitas los que m치s likes tienen? 

Para solucionar este ejercicio deber칠is:

> Agrupar por artista y calcular la media de visitas y de likes.


```python
df_sy.head()
```

```python
df_meanViews = df_sy.groupby('artist')['views'].mean().reset_index()
df_meanLikes = df_sy.groupby('artist')['likes'].mean().reset_index()
df_meanViews.head()
df_meanLikes.head()
```

> Quedaros con los 10 artistas que m치s visitas han tenido.


```python
df_sortedViews = df_meanViews.sort_values(by = 'views', ascending = False).head(10)
df_sortedViews
```

> Quedaros con los 10 artistas que m치s likes han tenido.


```python

#Vemos que no hay una correlaci칩n clara entre el n칰mero de vistas de los videos y las escuchas:
#Son dos formatos diferentes que atraen a diferentes tipos de p칰blico.

df_sortedLikes = df_meanLikes.sort_values(by = 'likes', ascending = False).head(10)
df_sortedLikes
```

> Haced un subplot, donde en uno tengamos los 10 artistas con m치s likes y en el otro los 10 artistas con mas views.


```python

fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20,15))

sns.barplot(x = df_sortedViews['artist'], 
             y = df_sortedViews['views'], 
             color = "#4374B3",
             ax = axes[0])

sns.barplot(x = df_sortedLikes['artist'],
             y = df_sortedLikes['likes'],
             color = "#69b3a2",
             ax = axes[1])

axes[0].set_xlabel('artists')
axes[0].set_ylabel('mean likes')
axes[1].set_xlabel('artists')
axes[1].set_ylabel('mean views')

axes[0].set_title("10 artitas con m치s LIKES", loc="center")
axes[1].set_title("10 artitas con m치s Visitas", loc="center")

plt.show()
```
